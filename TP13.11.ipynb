```r
# Load the necessary library
library(ggplot2)

# Generate sample data with exponentially increasing energy consumption over time
years <- seq(1970, 2024)
energy_consumption <- 2 * exp(0.05 * (years - 1970))  # Exponential growth formula

# Create a data frame with years and energy consumption
data <- data.frame(years = years, energy_consumption = energy_consumption)

# Plot the histogram using ggplot2
histogram_plot <- ggplot(data, aes(x = years, y = energy_consumption)) +
  geom_bar(stat = "identity", fill = "orange", alpha = 1) +  # Create bars with height corresponding to energy consumption
  labs(title = "World Energy Consumption (1970-2024)",  # Title of the plot
       x = "Year",  # X-axis label
       y = "Energy Consumption (in EJ)") +  # Y-axis label
  theme_minimal()  # Use a minimal theme for a cleaner look

# Print the plot
print(histogram_plot)
```
